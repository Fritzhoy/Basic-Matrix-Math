Step 1 - Create 3 vectors of 10 dimensions each, with random values considering the following ranges:

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Real values between [0,1)
vectorA = np.random.default_rng(seed=1).random(10)
vectorA
```

```python
# integer values between [50, 100)
vectorB = np.random.default_rng(seed=1).integers(50,100,10)
vectorB
```

```python
# real values between [50, 100)
vectorC = np.random.default_rng(seed=1)
vectorC = 100 - vectorC.random(10)*50
# vectorC = rng.random(10) + rng.integers(50,100,10) 
vectorC
```

Step 2 - Given the appropriate methods from NumPy, show from the comparison of results the following 

```python
# Commutative: A*B = B*A
A = np.round(vectorA,2)
B = np.round(vectorB,2)
AB = np.dot(A,B)
BA = np.dot(B,A)
# if AB == BA:
#     print(f'It is Commutative: A*B == B*A, AB = {AB}, BA = {BA}')
# else:
#     print(f'It is not Commutative: A*B == B*A, AB = {AB}, BA = {BA}') 
print(f'Commutative({AB == BA}): {AB} = {BA}')    
```

```python
# Distributive: A*(B+C) = A*B + A*C

A = vectorA
B = vectorB
C = vectorC
BC = vectorB + vectorC
ABC = np.sum(A*BC)
ABAC = ((np.sum(A*B)) + (np.sum(A*C)))
if ABC == ABAC:
    print(f'It is Distributive: A*(B+C) = {ABC},  A*B + A*C = {ABAC}')
else:
    print(f'It is not Distributive: A*(B+C) = {ABC},  A*B + A*C = {ABAC}') 
  
```

```python
# Not-associative: (A*B)*C ≠ A*(B*C)

A = vectorA
B = vectorB
C = vectorC
ABvectC = np.round(np.dot(np.dot(A,B), C),2)
vectABC = np.round(np.dot(A, np.dot(B,C)),2)

if np.array_equal(ABvectC,vectABC):
     print(f'It is Associative: (A*B)*C ≠ A*(B*C); (A*B)*C = {ABvectC}, A*(B*C)={vectABC}') 
else:
    print(f'It is Not-associative: (A*B)*C ≠ A*(B*C); (A*B)*C = {ABvectC}, A*(B*C)={vectABC}')
   
```

### Exercise 2

Create 2 vectors A and B with 3 dimensions each with elements of your choice. Create a third vector C that stores the Hadamard multiplication between A and B. Next, use the Matplotlib to present a 3D plot with 3 three vectors.

```python
import numpy as np
from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt

#activate the interactive plot
%matplotlib notebook

#Create the 3D axes
fig = plt.figure(figsize = (8,8))
ax = plt.axes(projection='3d')
ax.grid()

#Create Vector A and B, Vector C stores the Hadamard multiplication between A and B. 
vetorA = np.array([3, 2, 6])
vetorB = np.array([1, 5, 3])
vetorC = vetorA*vetorB


ax.plot3D([0,vetorA[0]], [0,vetorA[1]], [0,vetorA[2]],'k-', label='vetorA')
ax.plot3D([0,vetorB[0]], [0,vetorB[1]], [0,vetorB[2]],'m-', label='vetorB')
ax.plot3D([0,vetorC[0]], [0,vetorC[1]], [0,vetorC[2]],'r--', label='vetorA*vetorB')
ax.set_title('3D Vetor Plot')
ax.legend()

# Set axes label
ax.set_xlabel('Eixo x', labelpad=20)
ax.set_ylabel('Eixo y', labelpad=20)
ax.set_zlabel('Eixo z', labelpad=20)

plt.show()
```
